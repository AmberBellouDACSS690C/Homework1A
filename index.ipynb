```python
import geopandas as gpd

# Option A: Read from local file
gdf_binned = gpd.read_file("countriesCIA_binned.gpkg")
best_countries = gdf_binned[
    (gdf_binned["obesity_bin"] == 1) &
    (gdf_binned["alcohol_bin"] == 1) &
    (gdf_binned["tobacco_bin"] == 1)
]
best_countries
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,6))

# Plot all countries in light gray
gdf_binned.plot(ax=ax, color="lightgray", edgecolor="white")

# Overlay best countries in red
best_countries.plot(ax=ax, color="red")

plt.title("Countries with the Best (Lowest) Obesity, Alcohol, and Tobacco Rates", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
